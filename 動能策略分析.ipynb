```python
#套件更新  (備用，不需執行)
!pip install --upgrade pandas-datareader
!pip install --upgrade pandas

#import 與 設定
import datetime as t
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as pdr_web
import numpy as np
import yfinance as yf
yf.pdr_override()
from dateutil.relativedelta import relativedelta   #設定月日期的間隔差使用

# 設定開始與結束時間
start = t.datetime(2016,1,1)
end = t.datetime(2020,12,31)
```

```python
# 爬取wiki上的 Dow 30 清單
url = 'https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average'
df = pd.read_html(url)[1]
print(df.head(5))

symbol = df["Symbol"].to_list()
print(symbol)
```

```python
#爬取月收盤資料
stock_df = pdr_web.get_data_yahoo(symbol, start, end, interval='1mo')['Adj Close']

display(stock_df.head(5))  #展示頭五筆資料
display(stock_df.tail(5))  #展示末五筆資料
```

```python
# 計算當月報酬，並建立新的dataframe: month_ret
month_ret = stock_df.pct_change()  # 創造新的dataframe: month_ret

# 計算過去6個月報酬，並建立新的dataframe: past_6m_ret
past_6m_ret = (month_ret+1).rolling(6).apply(np.prod)-1

#未來6個月報酬
future_6m_ret = (month_ret+1).shift(-6).rolling(6).apply(np.prod)-1
```

```python
# 創一個 Dataframe 只存贏家和輸家的 6-month Returns
Mom_6M_ret = stock_df
Mom_6M_ret["Winner_6M_ret"] = np.nan  #指定新欄位，先存入NaN值
Mom_6M_ret["Loser_6M_ret"] = np.nan   #指定新欄位，先存入NaN值

Mom_6M_ret = Mom_6M_ret[["Winner_6M_ret", "Loser_6M_ret"]] #df其他的資料被清空了，只留兩個

#計算 "開始月" 與 "結束月" 的月份相差數
num_months = (end.year-start.year)*12 + (end.month-start.month) + 1
print(num_months)
```

```python
for num in range(6,num_months):
  form_month = start + relativedelta(months = num)#六個月後up的年月日
  ret_form = past_6m_ret.loc[form_month]#loc:只顯示form_month的前六月平均報酬
  ret_form = ret_form.reset_index()#重新設定索引
  ret_form['quintile'] = pd.qcut(ret_form[form_month], 5,labels=False)

  winners = ret_form[ret_form.quintile ==4]
  losers = ret_form[ret_form.quintile ==0]
  winners = winners.rename(columns={"index": "Symbols"})
  losers = losers.rename(columns={"index": "Symbols"})
  winner_ret_6m = future_6m_ret.loc[form_month , future_6m_ret.columns.isin(winners.Symbols)]#從 future_6m_ret 中選取贏家和輸家的未來6個月報酬率數據。
  loser_ret_6m = future_6m_ret.loc[form_month , future_6m_ret.columns.isin(losers.Symbols)]

  Mom_6M_ret.loc[form_month,"Winner_6M_ret"] = winner_ret_6m.mean()
  Mom_6M_ret.loc[form_month,"Loser_6M_ret"] = loser_ret_6m.mean()
  #將當前迭代中贏家和輸家的未來6個月平均報酬率分別存儲在 Mom_6M_ret DataFrame 中的相應列（"Winner_6M_ret" 和 "Loser_6M_ret")
```

```python
#展示剛剛算出的動能策略在樣本期間的基礎統計量
display(Mom_6M_ret.describe())

# 爬取與計算 Dow 30 Returns
stock_df_Dow30 = pdr_web.get_data_yahoo("^DJI", start, end, interval='1mo')['Adj Close'].pct_change()
Dow30_6m_ret = (stock_df_Dow30+1).shift(-6).rolling(6).apply(np.prod)-1

display(Dow30_6m_ret.describe())
```
